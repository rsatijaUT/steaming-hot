```python
s_scaler = StandardScaler()
mm_scaler = MinMaxScaler()
ma_scaler = MaxAbsScaler()
r_scaler = RobustScaler()
qt_scaler = QuantileTransformer()
pt_scaler = PowerTransformer(method='yeo-johnson')
```

```python
#Define a function to plot different scaling of data

def scale_data(df, games,cutoff=-100):
    df["mean"] = df.mean(axis=1)
    df = df.loc[df['mean']>cutoff].reset_index(drop=True)
    df_data = df.drop(columns=['real_id', 'name','mean','Last 30 Days']).copy()
    pd.to_datetime(df_data.columns)
    #Scale data
    s_scaler.fit(df_data)
    df_sscaler = s_scaler.transform(df_data)

    mm_scaler.fit(df_data)
    df_mmscaler = mm_scaler.transform(df_data)

    ma_scaler.fit(df_data)
    df_mascaler = ma_scaler.transform(df_data)

    r_scaler.fit(df_data)
    df_rscaler = r_scaler.transform(df_data)
    
    qt_scaler.fit(df_data)
    df_qtscaler = qt_scaler.transform(df_data)
    
    pt_scaler.fit(df_data)
    df_ptscaler = pt_scaler.transform(df_data)
    
    df_uvscaler = df_data.apply(lambda x:x/np.linalg.norm(x,1))

    #Plot scaled data
    fig, ax = plt.subplots(3,3,figsize=(20,14))

    for i in games:
        ax[0,0].plot(df_data.iloc[i][::-1]/(1e6),label=df.name[i])
    ax[0,0].legend() 
    ax[0,0].set_title('Original/10^6')

    for i in games:
        ax[0,1].plot(df_sscaler[i][::-1],label=df.name[i])
    ax[0,1].legend()
    ax[0,1].set_title('StandardScaler')

    for i in games:
        ax[0,2].plot(df_mascaler[i][::-1],label=df.name[i])
    ax[0,2].legend()
    ax[0,2].set_title('MaxAbsScaler')

    for i in games:
        ax[1,0].plot(df_data.div(df['mean'],axis=0).iloc[i][::-1],label=df.name[i])
    ax[1,0].legend() 
    ax[1,0].set_title('MeanScaled')

    for i in games:
        ax[1,1].plot(df_mmscaler[i][::-1],label=df.name[i])
    ax[1,1].legend()
    ax[1,1].set_title('MinMaxScaler')

    for i in games:
        ax[1,2].plot(df_rscaler[i][::-1],label=df.name[i])
    ax[1,2].legend()
    ax[1,2].set_title('RobustScaler')
    
    for i in games:
        ax[2,0].plot(df_qtscaler[i][::-1],label=df.name[i])
    ax[2,0].legend() 
    ax[2,0].set_title('Quantile Transformer Scaler')

    for i in games:
        ax[2,1].plot(df_ptscaler[i][::-1],label=df.name[i])
    ax[2,1].legend()
    ax[2,1].set_title('Power Transformer Scaler')

    #for i in games:
     #   ax[2,2].plot(df_uvscaler[i][::-1],label=df.name[i])
    #x[2,2].legend()
    #ax[2,2].set_title('Unit Vector Scaler')

    plt.show()
```

```python
avg_player["mean"] = avg_player.mean(axis=1)
avg_player = avg_player.loc[avg_player['mean']>100].reset_index(drop=True)
avg_player_data = avg_player.drop(columns=['real_id', 'name','mean']).copy()

#Scale data
s_scaler.fit(avg_player_data)
avg_player_sscaler = s_scaler.transform(avg_player_data)

mm_scaler.fit(avg_player_data)
avg_player_mmscaler = mm_scaler.transform(avg_player_data)

ma_scaler.fit(avg_player_data)
avg_player_mascaler = ma_scaler.transform(avg_player_data)

r_scaler.fit(avg_player_data)
avg_player_rscaler = r_scaler.transform(avg_player_data)


#Plot data
fig, ax = plt.subplots(2,3,figsize=(20,14))

games = range(4)

for i in games:
    ax[0,0].plot(avg_player_data.iloc[i][::-1],label=avg_player.name[i])
ax[0,0].legend() 
ax[0,0].set_title('Avg. no. of players')

for i in games:
    ax[0,1].plot(avg_player_sscaler[i][::-1],label=avg_player.name[i])
ax[0,1].legend()
ax[0,1].set_title('StandardScaled avg. no. of players')

for i in games:
    ax[0,2].plot(avg_player_mascaler[i][::-1],label=avg_player.name[i])
ax[0,2].legend()
ax[0,2].set_title('MaxAbsScaled avg. no. of players')

for i in games:
    ax[1,0].plot(avg_player_data.div(avg_player['mean'][::-1],axis=0).iloc[i],label=avg_player.name[i])
ax[1,0].legend() 
ax[1,0].set_title('MeanScaled avg. no. of players')

for i in games:
    ax[1,1].plot(avg_player_mmscaler[i][::-1],label=avg_player.name[i])
ax[1,1].legend()
ax[1,1].set_title('MinMaxScaled avg. no. of players')

for i in games:
    ax[1,2].plot(avg_player_rscaler[i][::-1],label=avg_player.name[i])
ax[1,2].legend()
ax[1,2].set_title('RobustScaled avg. no. of players')

plt.show()
```
