```python
from bs4 import BeautifulSoup
#from urllib.request import urlopen
import pandas as pd
import numpy as np
import requests

from sklearn.metrics.pairwise import pairwise_distances

#For plotting
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import datetime as dt

# https://towardsdatascience.com/all-about-feature-scaling-bcc0ad75cb35
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import PowerTransformer
```

```python
# We grab data from the stored csv files:
data = pd.read_csv('GamesDataAll.csv')

# Make percentile into number and clear "-"
data = data.replace(to_replace='%', value='', regex=True).replace({'-':'NaN'})#.replace(to_replace='Inf', value=+100, regex=True)
#data['Perc_Gain'] = data['Perc_Gain']/100

data.head()
```

```python
# Get Perc_Gain and change data to float.
p_gain = data.loc[data['Perc_Gain']==1]
p_gain = p_gain.drop(columns=['App_id', 'Name', 'Avg_Players', 'Gain', 'Perc_Gain', 'Peak_Players',
       'Last 30 Days']).astype(float)

p_gain.sample(5)
```

```python
# Get Avg_Players.
avg_players = data.loc[data['Avg_Players']==1]
avg_pure = avg_players.drop(columns=['App_id', 'Name', 'Avg_Players', 'Gain', 'Perc_Gain', 'Peak_Players',
       'Last 30 Days']).astype(float)

avg_pure.sample(5)
```

```python
avg_pure['max'] = avg_pure.max(axis=1)
avg_pure.sort_values(by=['max'], ascending=False, inplace=True)#.reset_index(drop=True)

avg_max = avg_pure['max'].max()+1
bins = [0,50,400,3300,avg_max]
counts  = avg_pure['max'].value_counts(bins=bins, sort=False) 
#plt.scatter(bins[:-1],counts)
print(counts)
print()
print('Using maximum of Avg_Players:')
for i in [100,500,2000,5000]:
    print('The cutoff for top '+str(i)+' games is '+str(avg_pure.reset_index(drop=True)['max'][i]))
```

```python
# Get Peak_Players.
peak_players = data.loc[data['Peak_Players']==1]
peak_pure = peak_players.drop(columns=['App_id', 'Name', 'Avg_Players', 'Gain', 'Perc_Gain', 'Peak_Players',
       'Last 30 Days']).astype(float)

peak_pure['max'] = peak_pure.max(axis=1)
peak_pure.sort_values(by=['max'], ascending=False, inplace=True)

peak_pure.sample(5)
```

```python
peak_max = peak_pure['max'].max()+1
bins = [0,50,400,4000,peak_max]
counts  = peak_pure['max'].value_counts(bins=bins, sort=False) 
#plt.scatter(bins[:-1],counts)
print(counts)
print()
print('Using maximum of Peak_Players:')
for i in [100,500,2000,5000]:
    print('The cutoff for top '+str(i)+' games is '+str(peak_pure.reset_index(drop=True)['max'][i]))
```

```python
s_scaler = StandardScaler()
mm_scaler = MinMaxScaler()
ma_scaler = MaxAbsScaler()
r_scaler = RobustScaler()
qt_scaler = QuantileTransformer()
pt_scaler = PowerTransformer(method='yeo-johnson')
```

```python
#Define a function to plot different scaling of data

scaler = []

def scale_data(df, games,cutoff=-100):
    df["mean"] = df.mean(axis=1)
    df = df.loc[df['mean']>cutoff].reset_index(drop=True)
    df_data = df.drop(columns=['mean','max']).copy()
    pd.to_datetime(df_data.columns)
    #Scale data
    s_scaler.fit(df_data)
    df_sscaler = s_scaler.transform(df_data)

    mm_scaler.fit(df_data)
    df_mmscaler = mm_scaler.transform(df_data)

    ma_scaler.fit(df_data)
    df_mascaler = ma_scaler.transform(df_data)

    r_scaler.fit(df_data)
    df_rscaler = r_scaler.transform(df_data)
    
    qt_scaler.fit(df_data)
    df_qtscaler = qt_scaler.transform(df_data)
    
    pt_scaler.fit(df_data)
    df_ptscaler = pt_scaler.transform(df_data)
    
    df_uvscaler = df_data.apply(lambda x:x/np.linalg.norm(x,1))

    #Plot scaled data
    fig, ax = plt.subplots(3,3,figsize=(20,14))

    for i in games:
        ax[0,0].plot(df_data.iloc[i][::-1]/(1e6),label=df.Name[i])
    ax[0,0].legend() 
    ax[0,0].set_title('Original/10^6')

    for i in games:
        ax[0,1].plot(df_sscaler[i][::-1],label=df.Name[i])
    ax[0,1].legend()
    ax[0,1].set_title('StandardScaler')

    for i in games:
        ax[0,2].plot(df_mascaler[i][::-1],label=df.Name[i])
    ax[0,2].legend()
    ax[0,2].set_title('MaxAbsScaler')

    for i in games:
        ax[1,0].plot(df_data.div(df['mean'],axis=0).iloc[i][::-1],label=df.Name[i])
    ax[1,0].legend() 
    ax[1,0].set_title('MeanScaled')

    for i in games:
        ax[1,1].plot(df_mmscaler[i][::-1],label=df.Name[i])
    ax[1,1].legend()
    ax[1,1].set_title('MinMaxScaler')

    for i in games:
        ax[1,2].plot(df_rscaler[i][::-1],label=df.Name[i])
    ax[1,2].legend()
    ax[1,2].set_title('RobustScaler')
    
    for i in games:
        ax[2,0].plot(df_qtscaler[i][::-1],label=df.Name[i])
    ax[2,0].legend() 
    ax[2,0].set_title('Quantile Transformer Scaler')

    for i in games:
        ax[2,1].plot(df_ptscaler[i][::-1],label=df.Name[i])
    ax[2,1].legend()
    ax[2,1].set_title('Power Transformer Scaler')

    #for i in games:
     #   ax[2,2].plot(df_uvscaler[i][::-1],label=df.Name[i])
    #x[2,2].legend()
    #ax[2,2].set_title('Unit Vector Scaler')

    plt.show()
```
